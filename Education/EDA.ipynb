DEFINE A FUNCTION TO EXTRACT THE TITLE, FIRST NAME AND LAST NAME

```python
def extract_name_components(name):
    #Split the name by comma to separate the last name from the rest
    last_name,rest_of_name=name.split(',',1)


    #split the rest of the name by spaces to separate title and first name
    name_parts=rest_of_name.strip().split(' ')

    #Extract title
    title=name_parts[0]

    #First name might have multiple parts,so join them except title
    first_name=' '.join(name_parts[1:])

    #Clean up any extra characters from first name
    first_name= first_name.replace('(','').replace(')','')

    return title,first_name,last_name.strip()
```

```python
#Apply the function to the name column
data['Title'],data['First Name'],data['Last Name']=zip(*data['Name'].map(extract_name_components))
data.head(5)
```

```python
data.drop('Name',axis=1,inplace=True)
data.head(5)
```
