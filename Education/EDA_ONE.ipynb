```python
# Removes € sign
def remove_suffix(value):
  if value.startswith('€'):
    value = value[1:] 

  if value[-1].isdigit(): # no suffix
    return float(value)
  # handle cases where there's suffix (M or K)
  elif value[-1] == "M":
    return float(value[:-1]) * 10**6
  elif value[-1] == "K":
    return float(value[:-1]) * 10 ** 3
  else:
    return None
```

```python
print(remove_suffix("€100K"))
print(remove_suffix("€100K"))
print(remove_suffix("€6M"))
```

```python
Fifa[["Value", "Wage", "Release Clause"]] = Fifa[["Value", "Wage", "Release Clause"]].applymap(remove_suffix)
Fifa[["Value", "Wage", "Release Clause"]]
```
