IN PYTHON, A LOOP IS A CONTROL STRUCTURE THAT ALLOWS YOU TO EXECUTE A BLOCK OF CODE REPEATEDLY BASED ON CERTAIN CONDITIONS

FOR LOOP IS LIKE SAYING, "DO SOMETHING A SPECIFIC NUMBER OF TIMES." IT REPEATS THE SAME TASK FOR A SET NUMBER OF TIMES, AND YOU USUALLY KNOW IN ADVANCE HOW MANY TIMES IT WILL RUN.

FOR LOOP IS USED TO ITERATE OVER A SEQUENCE OF ITEMS. 

```python
for i in range(3,10):
    print(i)
```

```python
for z in range(0,201,20):
    print(z)
```

```python
for z in range(0,201,20):
    if z<50:
        continue
    print(z)
```

```python
for z in range(0,201,10):
    if z==50:
        break
    print(z)
```

```python
for u in range(10):
    print(u)
```

A WHILE LOOP IS LIKE SAYING, "DO SOMETHING AS LONG AS A CONDITION IS TRUE." IT KEEPS REPEATING THE TASK UNTIL A CONDITION IS NO LONGER TRUE.

WHILE LOOP IS USED TO REPEATEDLY EXECUTE A BLOCK OF STATEMENTS WHILE A CONDITION IS TRUE.

```python
X=0
while X<5:
    print(X)
    X+=1
```

```python
x=5
iters=0
while iters<x:
    print("Princewill")
    iters+=1
```
